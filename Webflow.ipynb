```python
headers = {
        'accept-encoding': 'gzip, deflate, sdch, br',
        'accept-language': 'en-US,en;q=0.8,zh-CN;q=0.6,zh;q=0.4',
        'upgrade-insecure-requests': '1',
        'user-agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/56.0.2924.87 Safari/537.36',
        'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'cache-control': 'max-age=0',
    }

url = 'https://forum.webflow.com/'
request = requests.get(url,headers)
soup = BeautifulSoup(request.text,'lxml')
```

```python
title = soup.title
print(title)
```

```python
title_class = soup.find('div',{'class':'wrap'})
print(title_class)
```

```python
title_class2 = title_class.find('span',{'class':'link-top-line'})
print(title_class2)
```

```python
title_text = title_class2.getText()
print(title_text)
```

```python
title_tag = title_class2.find('a', {'class':'title raw-link raw-topic-link'})
print(title_tag)
```

```python
cat_tag = title_class.find('div',{'class':'link-bottom-line'}).findAll('span',{'class':'badge-category clear-badge'})
print(cat_tag)
```

```python
cat = cat_tag[0].getText()
print(cat)
```
